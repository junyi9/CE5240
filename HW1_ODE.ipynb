```python
import numpy as np
from scipy.integrate import solve_ivp

def Vlead(t):
    if t < 5:
        vl = 20 - 10 * (t / 5) ** 3
    else:
        vl = 10
    return vl

def system(t, Y):
    C = np.array([[0, -1], [0.23, -0.323]])
    D = np.array([[0, 1], [1, 0.07]])
    vl = Vlead(t)
    return C @ Y + D @ np.array([-1.61, vl])

tspan = [0, 20]
y0 = [29, 20]  # initial conditions
sol = solve_ivp(system, tspan, y0, t_eval=np.linspace(tspan[0], tspan[1], 200))
```

```python
import matplotlib.pyplot as plt
plt.plot(sol.y[0])
plt.show()
```
